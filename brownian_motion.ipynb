## Simple brownian motion



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean=0#drift
variance=dt#volatality

T = 1.0  # Total time
N = 1000  # Number of time steps
dt = T / N  # step size
num_paths = 5  # Number of sample paths
t = np.linspace(0, T, N+1)
W = np.zeros((N+1, num_paths))
for i in range(1, N+1):
    dW = np.random.normal(0, np.sqrt(dt), size=num_paths)
    W[i] = W[i-1] + dW
```

# Geometric Brownian Motion

### here we can see in above plot that w(t) is also giving negative values but we know price of a stock can't be negative . So to avoid this we use geometric brownian motion.

```python
Y=np.zeros((N+1,num_paths))
for i in range(1, N+1):
    dW = np.random.normal(0, np.sqrt(dt), size=num_paths)
    W[i] = W[i-1] + dW
    Y[i]=np.exp(W[i])

plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
for j in range(num_paths):
    plt.plot(t, W[:, j])
plt.xlabel("time")
plt.ylabel("$W_t$")
plt.title("simple brownian motion")

plt.subplot(2,2,3)
for j in range(num_paths):
  plt.plot(t,Y[:,j])
plt.xlabel("time")
plt.ylabel("$W_t$")
plt.title("geometric brownian motion")
plt.show()



```
